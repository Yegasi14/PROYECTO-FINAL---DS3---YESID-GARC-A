##**Creación de columna Sentimiento y metadata**

En esta sección queremos crear una clasificación rapida de sentimiento basado en la siguiente lógica:

Si el Rating es de 1 a 3, el sentimiento es 0 y por tanto **MALO**, por el contrario si el el Rating está entre 4 y 5 pede considerarse como **BUENO**.

Esto nos ayudará en el proyecto a dar manejo a un modelo de clasificación mas simple si se necesita como por ejemplo una regresión logistica que se basa en una sigmoide y por tanto es más facil de evaluar.

```python
# Define una función para clasificar la calificación
def clasificar_sentimiento(calificacion):
    if calificacion <= 3:
        return 0  # Malo
    else:
        return 1  # Bueno

# Aplica la función para crear una nueva columna 'Feeling'
df['Feeling'] = df['Rating'].apply(clasificar_sentimiento)
# df.drop('Sentimiento', axis=1, inplace=True) # Eliminada esta línea

# Crea una nueva columna 'Score' con los valores de texto
df['Score'] = df['Feeling'].apply(lambda x: 'GOOD' if x == 1 else 'BAD')

# Muestra las primeras filas con la nueva columna
display(df.head())
```

### **Preprocesamiento basico de textos**

Los siguientes pasos permiten construir una nueva columna llamada clean_review, con una versión simplificada y normalizada del texto original, libre de signos de puntuación y palabras vacías, lo cual dará mayor precisión a la hora de aplicar nuestra tecnica de NLP.

###**Tokenización**

```python
if 'clean_review' not in df.columns:
    # Asegurarse de que los recursos de NLTK estén descargados
    try:
        nltk.data.find('tokenizers/punkt')
    except LookupError:
        nltk.download('punkt')

    try:
        nltk.data.find('corpus/stopwords')
    except LookupError:
        nltk.download('stopwords')

    try:
        nltk.data.find('corpora/wordnet')
    except LookupError:
        nltk.download('wordnet')


    tokenizer = TreebankWordTokenizer()
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()

    def limpiar_texto(texto):
        # Asegurarse de que el texto sea una cadena antes de procesar
        texto = str(texto)
        # Eliminar caracteres no alfanuméricos y convertir a minúsculas
        texto = re.sub(r'[^\w\s]', '', texto.lower())
        # Tokenizar
        tokens = tokenizer.tokenize(texto)
        # Eliminar stop words
        tokens = [t for t in tokens if t not in stop_words]
        # Lematizar
        lemmas = [lemmatizer.lemmatize(t) for t in tokens]
        return " ".join(lemmas)

    # Aplicar la función de limpieza a la columna 'Review'
    df['clean_review'] = df['Review'].fillna('').apply(limpiar_texto)

# Mostrar las primeras filas con la nueva columna
display(df[['Review','clean_review']].head(10))
```

Esta transformación llena la columna clean_review con texto ya procesado. Hemos aplicado varias técnicas clave de procesamiento de lenguaje natural, que incluyen:

* **Conversión a minúsculas:** Estandariza todo el texto.

* **Eliminación de caracteres especiales:** Remueve el ruido no alfanumérico.

* **Filtrado de palabras vacías (stop words):** Elimina palabras comunes y menos significativas (por ejemplo, "el", "un", "es").

* **Lematización:** Reduce las palabras a su forma base o raíz.

Esta limpieza exhaustiva asegura que los datos estén preparados para un análisis gramatical robusto usando spaCy, lo que a su vez mejora la precisión y relevancia de las características utilizadas en nuestros modelos.

###**Conteo de POS**

Esto permite entender la composición lingüística de las reseñas: si predominan sustantivos (descripciones de aspectos del hotel, como habitaciones, servicios o instalaciones), adjetivos (valoraciones), verbos (acciones o experiencias), o adverbios (intensidad o matices).

Este patrón permite inferir la riqueza descriptiva y a anticipar cómo los clasificadores pueden detectar el sentimiento.

###**Extracción de entidades nombradas (NER)**

En esta sección, se implementa la técnica de Reconocimiento de Entidades Nombradas (NER) utilizando la librería spaCy sobre una muestra del corpus de reseñas preprocesado.

El objetivo principal es la extracción y categorización automatizada de entidades clave como nombres de hoteles (ORG), ubicaciones geográficas (GPE), fechas (DATE) y nombres propios (PERSON).

La aplicación de NER permite transformar el texto no estructurado en datos estructurados. Esto no solo facilita la detección de patrones de mención, sino que también permite un análisis de tópicos más preciso y una segmentación granular de las opiniones, lo que es crucial para comprender qué aspectos específicos de la experiencia (hoteles, destinos, fechas de estancia) son más relevantes para la percepción del sentimiento del cliente.

###**N‑gramas (bigrama/trigrama)**

Esta sección se centra en la ingeniería de características léxicas mediante el análisis de n-gramas. Los n-gramas son secuencias contiguas de n palabras que capturan colocaciones y patrones de co-ocurrencia en el texto. A diferencia de un simple análisis de palabras individuales, la identificación de n-gramas permite retener el contexto y el significado compuesto de frases, lo que es fundamental para el análisis de sentimientos.

Para esta tarea, el corpus de reseñas es previamente procesado con spaCy para la tokenización y lematización, lo que asegura que las palabras se encuentren en su forma base (p. ej., "limpio" en lugar de "limpieza") antes de la extracción de n-gramas. A continuación, se emplea la clase CountVectorizer de scikit-learn para crear una matriz de características. Esta herramienta es ideal, ya que permite definir el rango de n-gramas a considerar (p. ej., ngram_range=(2, 3) para bigramas y trigramas) y aplicar un filtro de frecuencia mínima (min_df) para eliminar n-gramas raros y no relevantes, asegurando que solo se incluyan en el análisis los patrones más significativos y recurrentes.

La relevancia de este enfoque radica en su capacidad para capturar el sentimiento a nivel de frase. Por ejemplo, un modelo que solo analiza palabras individuales podría interpretar "service" y "poor" como términos neutros o de baja polaridad por separado, pero la combinación "poor service" es un claro indicador de sentimiento negativo. De manera similar, frases como "friendly staff" o "clean room" son esenciales para identificar experiencias positivas. En la fase de modelado, estos n-gramas se transformarán en características que mejoran significativamente el poder predictivo de los algoritmos de clasificación de sentimiento.

```python
corpus = df['clean_review'].dropna().tolist()

def top_ngrams(corpus, ngram_range=(2,2), topk=20, min_df=2):
    vec = CountVectorizer(ngram_range=ngram_range, min_df=min_df)
    X = vec.fit_transform(corpus)
    freqs = zip(vec.get_feature_names_out(), X.sum(axis=0).A1)
    out = pd.DataFrame(sorted(freqs, key=lambda x: x[1], reverse=True)[:topk],
                       columns=['ngram','freq'])
    return out

# Calcular top bigramas y trigramas
top_bi  = top_ngrams(corpus, (2,2), topk=20)
top_tri = top_ngrams(corpus, (3,3), topk=15)

# Visualización de Top 20 bigramas con Plotly Express
fig_bi = px.bar(top_bi, x='freq', y='ngram', orientation='h',
                title='<b>Top 20 Bigramas</b>',
                labels={'freq': 'Frecuencia', 'ngram': 'Bigrama'},
                color_discrete_sequence=px.colors.sequential.Blues_r)
fig_bi.update_layout(yaxis={'categoryorder':'total ascending'})
fig_bi.update_layout(title_x=0.5)
fig_bi.show()

# Visualización de Top 15 trigramas con Plotly Express
fig_tri = px.bar(top_tri, x='freq', y='ngram', orientation='h',
                 title='<b>Top 15 Trigramas</b>',
                 labels={'freq': 'Frecuencia', 'ngram': 'Trigrama'},
                 color_discrete_sequence=px.colors.sequential.Viridis)
fig_tri.update_layout(yaxis={'categoryorder':'total ascending'})
fig_tri.update_layout(title_x=0.5)
fig_tri.show()
```

Los gráficos presentados visualizan las combinaciones de palabras más frecuentes y significativas en el corpus de reseñas de hoteles. Este enfoque es crucial para la extracción de características para modelos de clasificación de sentimiento.

Bigramas (n=2): Se identifican patrones léxicos de alta frecuencia que corresponden a aspectos clave de la experiencia del huésped. Expresiones como "friendly staff", "great location" y "clean room" no solo indican la polaridad del sentimiento, sino que también señalan directamente los tópicos más relevantes para los usuarios. Estos bigramas actúan como características potentes para un modelo de aprendizaje supervisado, ya que su presencia tiene un alto valor predictivo.

Trigramas (n=3): El análisis de trigramas ofrece una granularidad aún mayor en el contexto. Frases como "staff friendly helpful" o "hotel great location" brindan una descripción más rica y detallada de la opinión. Estas combinaciones más largas son particularmente valiosas en modelos de clasificación automática, ya que capturan matices y especificidades que pueden ser cruciales para discriminar entre opiniones matizadas (p. ej., "the room was fine but the staff was not very friendly").

En resumen, la identificación de estos patrones de lenguaje recurrentes a través del análisis de n-gramas confirma que las reseñas contienen una estructura de datos apta para el aprendizaje automático. La inclusión de bigramas y trigramas como características en la fase de entrenamiento del modelo mejora significativamente su capacidad para predecir y clasificar con precisión el sentimiento, ya que estos grupos de palabras codifican información semántica y contextual que no está presente en un simple modelo de bolsa de palabras (bag-of-words).

#**9. (VADER) ETIQUETADO DEBIL DE SENTIMIENTO**

En esta etapa del pipeline de procesamiento, se implementa una estrategia de etiquetado débil para asignar polaridad de sentimiento a cada reseña. Se utiliza el modelo VADER (Valence Aware Dictionary and sEntiment Reasoner), una herramienta léxica y de reglas heurísticas diseñada para el análisis de sentimientos en textos de redes sociales y foros.

VADER proporciona un puntaje de sentimiento compuesto y continuo, que oscila entre -1 (polaridad negativa extrema) y +1 (polaridad positiva extrema). Este puntaje es la base para la discretización del sentimiento. Se aplican umbrales predefinidos, donde los puntajes por debajo de -0.05 se clasifican como negativos, los puntajes por encima de +0.05 se clasifican como positivos, y los puntajes entre estos umbrales se consideran neutrales. .

Este conjunto de datos etiquetado de forma heurística se denomina "ground truth débil". Sirve como una referencia inicial para el entrenamiento y la evaluación de modelos de aprendizaje supervisado posteriores, como un clasificador clásico de baseline o una red neuronal como el Perceptrón Multicapa (MLP). El uso de VADER permite obtener un conjunto de datos etiquetado sin la necesidad de anotación manual, lo que es crucial para la puesta en marcha inicial de los modelos de clasificación.

```python
try:
    nltk.data.find('sentiment/vader_lexicon.zip')
except LookupError:
    nltk.download('vader_lexicon')

# Inicializar VADER
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Prevenir problemas con nulos en la columna 'Review'
df['Review'] = df['Review'].fillna('')

#Función para calcular puntaje compuesto
def score_sent(texto):
    return sia.polarity_scores(texto)['compound'] if isinstance(texto, str) else 0.0

# Aplicar la función para crear la columna 'sent_score'
df['sent_score'] = df['Review'].apply(score_sent)

#Función para convertir puntaje a etiqueta
def to_label(s):
    if s >= 0.05:
        return 'positive'
    elif s <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Aplicar la función para crear la columna 'Sentiment'
df['Sentiment'] = df['sent_score'].apply(to_label)

# Distribución absoluta
dist_abs = df['Sentiment'].value_counts().reindex(['negative','neutral','positive'])
dist_abs = dist_abs.reset_index()
dist_abs.columns = ['Sentiment', 'Count']


# Gráfico de barras con Plotly Express
fig = px.bar(dist_abs, x='Sentiment', y='Count',
             title='<b>Distribución de Sentimientos según VADER</b>',
             color='Sentiment',
             color_discrete_map={'negative': '#d62728', 'neutral': '#7f7f7f', 'positive': '#2ca02c'})

# Añadir porcentajes como texto en el gráfico
total = dist_abs['Count'].sum()
dist_abs['Percentage'] = (dist_abs['Count'] / total * 100).round(1)

fig.update_traces(texttemplate='%{customdata:.1f}%', textposition='outside', customdata=dist_abs['Percentage'])


fig.update_layout(title_x=0.5)
fig.show()

# Mostrar tabla porcentual
print("\nDistribución porcentual de sentimientos (%):")
display(dist_abs[['Sentiment', 'Percentage']])
```

Podemos observar que hay un claro desbalance de clases, con una gran mayoría de reseñas clasificadas como positivas (aproximadamente 91.9%), mientras que las categorías neutral y negativa tienen porcentajes mucho menores (alrededor del 0.4% y 7.7% respectivamente). Este desbalance es importante tenerlo en cuenta para las etapas posteriores del proyecto.

Si bien VADER es útil para obtener una clasificación inicial, su enfoque basado en léxicos puede tener limitaciones, especialmente con sarcasmo o expresiones más complejas que no estén en su diccionario. Sin embargo, para este conjunto de datos de reseñas de hoteles, donde el lenguaje tiende a ser más directo en cuanto a la satisfacción o insatisfacción, VADER parece ser una herramienta de etiquetado débil razonablemente efectiva para una primera aproximación.

La existencia de este desbalance de clases implica que, al entrenar modelos de clasificación supervisada, será crucial implementar estrategias para manejarlo. Esto puede incluir técnicas como el sobremuestreo (oversampling) de las clases minoritarias (negativas y neutrales) o el submuestreo (undersampling) de la clase mayoritaria (positiva) durante la división de los datos de entrenamiento y prueba. También será fundamental evaluar el rendimiento del modelo utilizando métricas apropiadas para datos desbalanceados, como precisión (precision), exhaustividad (recall) y F1-score, en lugar de depender únicamente de la exactitud general (accuracy), que puede ser engañosa en estos casos.

Con estas etiquetas de sentimiento generadas por VADER como nuestra "verdad fundamental débil", el siguiente paso lógico en el proyecto sería preparar los datos textuales para el modelado de aprendizaje automático, lo que implica la vectorización del texto (convertir las reseñas en representaciones numéricas) y posteriormente entrenar un modelo de clasificación de sentimientos supervisado.



# **10. MODELO TF-IDF PARA EL ENTRENAMIENTO DE LOS SENTIMIENTOS**

## **Entrenamiento y evaluación del modelo**

En esta sección se implementa un enfoque de clasificación de sentimiento basado en técnicas clásicas de Procesamiento de Lenguaje Natural (NLP) y Machine Learning.

Se utilizan los textos preprocesados de las reseñas de hoteles para construir representaciones vectoriales mediante TF-IDF (Term Frequency-Inverse Document Frequency). Esta técnica pondera la importancia de los términos (considerando unigramas y potencialmente bigramas) en función de su frecuencia en una reseña y su rareza en el corpus completo.

Posteriormente, se entrena un modelo de Regresión Logística, un algoritmo de clasificación lineal robusto y eficiente, utilizando las representaciones TF-IDF como características de entrada. Este modelo servirá como una línea base (baseline) para evaluar el rendimiento de enfoques más complejos, como el modelo de Deep Learning (MLP) que se desarrollará más adelante.

La evaluación técnica del modelo incluirá métricas estándar de clasificación para medir su desempeño predictivo, tales como la exactitud (accuracy) general, así como precisión (precision), exhaustividad (recall) y puntaje F1 (F1-score) por cada clase de sentimiento (positivo/negativo), además de la matriz de confusión para visualizar la distribución de predicciones correctas e incorrectas.

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
# Usar 'Sentiment' como base para la estratificación
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_review'], df['Sentiment'], test_size=0.2, random_state=42, stratify=df['Sentiment'])

# Inicializar y aplicar TF-IDF Vectorizer
# max_features: limitar el número de términos
# ngram_range: considerar unigramas y bigramas
# min_df: ignorar términos que aparecen en menos de 2 documentos
tfidf = TfidfVectorizer(max_features=20000, ngram_range=(1,2), min_df=2)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf  = tfidf.transform(X_test)

# Inicializar y entrenar el modelo de Regresión Logística
# max_iter: aumentar si la convergencia no se alcanza
# n_jobs: usar todos los núcleos de CPU disponibles
lr = LogisticRegression(max_iter=200, n_jobs=-1)
lr.fit(X_train_tfidf, y_train)

# Realizar predicciones en el conjunto de prueba
pred_lr = lr.predict(X_test_tfidf)

# Evaluar el modelo
print("Accuracy (LogReg):", round(accuracy_score(y_test, pred_lr), 3))
print("\nClassification Report (LogReg):\n", classification_report(y_test, pred_lr, digits=3))

# Mostrar matriz de confusión
cm = confusion_matrix(y_test, pred_lr, labels=['negative','neutral','positive'])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['negative','neutral','positive'],
            yticklabels=['negative','neutral','positive'])
plt.xlabel('Predicho'); plt.ylabel('Real'); plt.title('Matriz de confusión – TF‑IDF + LogReg')
plt.show()
```

## **Validación cruzada del modelo escogido**

Este enfoque de validación cruzada proporciona una estimación más confiable y robusta del rendimiento del modelo que una simple división única de datos de entrenamiento/prueba. En lugar de entrenar y evaluar el modelo una sola vez, este método lo hace k veces (en este caso, 5), utilizando un subconjunto diferente de los datos de entrenamiento para la validación en cada iteración.

* **Minimiza el sesgo de la muestra:** Una división aleatoria simple podría, por casualidad, crear un conjunto de prueba que no es representativo de todos los datos. La validación cruzada mitiga este riesgo al garantizar que cada porción del conjunto de entrenamiento sea utilizada tanto para el entrenamiento como para la validación.

* **Evalúa la estabilidad del modelo:** Al observar las variaciones en las métricas (precisión y F1) en cada fold, puedes ver cuán estable es el rendimiento de tu modelo. Un pequeño rango de variación (desviación estándar baja) indica un modelo más confiable.

```python
# --- Configuración de la validación cruzada ---
# N_SPLITS = 5 es un valor común, significa que dividiremos los datos en 5 'folds'
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

# Listas para almacenar las métricas de cada fold
accuracy_scores = []
f1_scores = []

# --- Inicializar el vectorizador TF-IDF ---
# Lo inicializamos aquí para que se ajuste en cada fold
tfidf_vectorizer = TfidfVectorizer(max_features=20000, ngram_range=(1,2), min_df=2)

print(f"Iniciando la validación cruzada K-Fold (k={n_splits}) para Regresión Logística...")

# Iterar sobre cada fold
for fold, (train_index, test_index) in enumerate(skf.split(X_train_feeling, y_train_feeling)):
    print(f"\n--- Fold {fold+1}/{n_splits} ---")

    # Dividir los datos del conjunto de entrenamiento en sub-entrenamiento y validación para este fold
    X_train_fold, X_val_fold = X_train_feeling.iloc[train_index], X_train_feeling.iloc[test_index]
    y_train_fold, y_val_fold = y_train_feeling.iloc[train_index], y_train_feeling.iloc[test_index]

    # Ajustar y transformar los datos de texto para este fold
    # Esto evita el 'data leakage' ya que el vectorizador aprende del subconjunto de entrenamiento
    X_train_tfidf_fold = tfidf_vectorizer.fit_transform(X_train_fold)
    X_val_tfidf_fold = tfidf_vectorizer.transform(X_val_fold)

    # Entrenar el modelo de Regresión Logística en el sub-conjunto de entrenamiento
    lr_model = LogisticRegression(max_iter=200, n_jobs=-1)
    lr_model.fit(X_train_tfidf_fold, y_train_fold)

    # Hacer predicciones en el sub-conjunto de validación
    predictions = lr_model.predict(X_val_tfidf_fold)

    # Calcular y almacenar las métricas para este fold
    fold_accuracy = accuracy_score(y_val_fold, predictions)
    fold_f1 = f1_score(y_val_fold, predictions, average='weighted')

    accuracy_scores.append(fold_accuracy)
    f1_scores.append(fold_f1)

    print(f"Precisión del Fold: {fold_accuracy:.4f}")
    print(f"Puntuación F1 del Fold: {fold_f1:.4f}")

# Calcular y mostrar las métricas promediadas de todos los folds
print("\n--- Resultados Promedio de la Validación Cruzada ---")
print(f"Precisión Promedio: {np.mean(accuracy_scores):.4f} (+/- {np.std(accuracy_scores):.4f})")
print(f"Puntuación F1 Promedio: {np.mean(f1_scores):.4f} (+/- {np.std(f1_scores):.4f})")
```

Los resultados de la validación cruzada K-Fold para el modelo de Regresión Logística muestran un rendimiento muy bueno y consistente.  A continuación, se presenta un análisis detallado y el siguiente paso a seguir.

El análisis de los resultados revela lo siguiente:

* **Rendimiento Promedio:** La precisión promedio del modelo es de 0.8988, lo que significa que, en promedio, el modelo clasifica correctamente casi el 90% de las revisiones de clientes. La puntuación F1 promedio de 0.8941 es casi idéntica, lo que indica un excelente equilibrio entre la precisión y la exhaustividad del modelo.

* **Consistencia:** La desviación estándar (el valor después del +/-) es de 0.0040 para la precisión y 0.0043 para la puntuación F1. Estos valores son muy bajos, lo que demuestra que el rendimiento del modelo es extremadamente consistente en los cinco "folds" o subconjuntos de datos. No hay un "fold" que se haya comportado significativamente peor que los demás, lo que sugiere que el modelo no se vio sesgado por un subconjunto particular de los datos de entrenamiento.

#**12. MODELO DE DEEP LEARNING: MLP (Keras/TensorFlow)**

En esta sección, se entrena una red neuronal multicapa (MLP) utilizando los vectores TF-IDF del texto como entrada. El objetivo es investigar si una arquitectura no lineal, con capas densas, una función de activación no lineal como ReLU y regularización por Dropout, es capaz de capturar combinaciones de términos que el modelo lineal de Regresión Logística no logra.

**Procedimiento Técnico Detallado:**
* **Codificación de Etiquetas:** Las etiquetas de sentimiento (positivas/negativas) se convierten a un formato numérico one-hot para que el modelo MLP las pueda procesar.

* **Definición de la Arquitectura:** Se construye una MLP simple con dos capas densas y se incorpora Dropout para regularizar el modelo y prevenir el sobreajuste.

* **Mecanismo de Parada Temprana:** Se implementa EarlyStopping con el objetivo de monitorear la métrica de validación durante el entrenamiento. Esto detiene el proceso de entrenamiento cuando el rendimiento del modelo en los datos de validación deja de mejorar, evitando el sobreajuste.

* **Evaluación y Comparación:** Finalmente, se evalúa el rendimiento del modelo entrenado en el conjunto de datos de prueba (test set) y se comparan los resultados con los del modelo de referencia (baseline) de Regresión Logística.

Esta prueba finaliza el objetivo central del proyecto: aplicar una técnica de Deep Learning a un problema de clasificación de sentimiento y analizar de manera rigurosa si esta arquitectura no lineal aporta una mejora tangible y real al rendimiento del modelo en comparación con el modelo lineal de Regresión Logística que ya se ha optimizado.

## **Adaptación y preparación de etiquetas para el modelo de Deep Learning**

**Nota:** Si las etiquetas ya están en un formato adecuado, esta parte puede no ser necesaria.
Sin embargo, si trabajamos con 3 o más etiquetas ('positive', 'negative', 'neutral'), to_categorical es fundamental para el 'one-hot encoding'.

```python
# Codificación de etiquetas. Esto convierte 'positive', 'negative', 'neutral' en 0, 1, 2, respectivamente.
# Luego, 'to_categorical' las convierte en vectores binarios (ej. 'positive' -> [1, 0, 0]).
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
y_train_oh = to_categorical(y_train_enc)
y_test_oh = to_categorical(y_test_enc)

# Definición de la dimensión de entrada para el modelo MLP
input_dim = X_train_tfidf.shape[1]

# Establecer la semilla para reproducibilidad
tf.random.set_seed(42)

# Construcción del modelo MLP usando Keras Sequential API
model = Sequential([
    # Primera capa densa con 256 neuronas y activación ReLU. 'input_shape' define la forma de la entrada.
    Dense(256, activation='relu', input_shape=(input_dim,)),
    # Capa de Dropout para regularización, evita el sobreajuste "apagando" aleatoriamente el 30% de las neuronas.
    Dropout(0.3),
    # Segunda capa densa con 128 neuronas y activación ReLU.
    Dense(128, activation='relu'),
    # Segunda capa de Dropout.
    Dropout(0.3),
    # Capa de salida. El número de neuronas es igual a la cantidad de clases.
    # La activación 'softmax' es ideal para problemas de clasificación multiclase.
    Dense(y_train_oh.shape[1], activation='softmax')
])

# Compilación del modelo: se define el optimizador, la función de pérdida y las métricas a monitorear.
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Implementación de EarlyStopping para detener el entrenamiento si la validación no mejora.
early = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Entrenamiento del modelo
history = model.fit(X_train_tfidf.toarray(), y_train_oh,
                    validation_split=0.2, # 20% del conjunto de entrenamiento se usa para validación.
                    epochs=15,            # Número de iteraciones completas sobre el conjunto de datos.
                    batch_size=64,        # Número de muestras por actualización de gradiente.
                    callbacks=[early],    # Se agrega el callback de EarlyStopping.
                    verbose=1)            # Muestra el progreso del entrenamiento.

# ---
### **Visualización de la Pérdida y Exactitud con Plotly**

# Creación de los gráficos de Pérdida y Exactitud para evaluar el rendimiento.
# Esto es crucial para identificar sobreajuste o bajoajuste.

fig = go.Figure()
# Gráfico de Pérdida
fig.add_trace(go.Scatter(y=history.history['loss'], name='Pérdida de Entrenamiento'))
fig.add_trace(go.Scatter(y=history.history['val_loss'], name='Pérdida de Validación'))
fig.update_layout(title='Curvas de Pérdida del Modelo', xaxis_title='Época', yaxis_title='Pérdida')
fig.show()

# Gráfico de Exactitud
fig = go.Figure()
fig.add_trace(go.Scatter(y=history.history['accuracy'], name='Exactitud de Entrenamiento'))
fig.add_trace(go.Scatter(y=history.history['val_accuracy'], name='Exactitud de Validación'))
fig.update_layout(title='Curvas de Exactitud del Modelo', xaxis_title='Época', yaxis_title='Exactitud')
fig.show()

# ---
### **Evaluación y Reporte de Resultados**

# Predicción en el conjunto de prueba
probs = model.predict(X_test_tfidf.toarray())
# np.argmax se usa para obtener el índice de la clase con mayor probabilidad.
# le.inverse_transform revierte la codificación numérica a las etiquetas originales.
pred_mlp = le.inverse_transform(np.argmax(probs, axis=1))

# Impresión de métricas clave.
print("Accuracy (MLP):", round(accuracy_score(y_test, pred_mlp), 3))
print(classification_report(y_test, pred_mlp, digits=3))


# Se crea y visualiza la matriz de confusión para entender el rendimiento del modelo en cada clase.
cm = confusion_matrix(y_test, pred_mlp, labels=['negative','neutral','positive'])
labels = ['negative','neutral','positive']
cm_df = go.Figure(data=go.Heatmap(
    z=cm,
    x=labels,
    y=labels,
    colorscale='Blues', # La paleta de colores se cambia a azul.
    text=cm,
    texttemplate="%{text}",
    hoverinfo="z"
))
cm_df.update_layout(
    title='Matriz de Confusión – MLP (TF‑IDF)',
    xaxis_title='Predicho',
    yaxis_title='Real',
    yaxis_autorange='reversed'
)
cm_df.show()
```

**Pérdida y Exactitud**
La primera visualización muestra las curvas de pérdida (loss) y exactitud (accuracy) a lo largo de las épocas de entrenamiento.

Esta versión optimizada aumenta la tasa de Dropout y añade regularización L2 (kernel_regularizer) a las capas densas. Esto ayuda a combatir el sobreajuste y a que el modelo generalice mejor con datos no vistos. Además, se mantienen las visualizaciones con Plotly para un análisis interactivo de los resultados.

```python
### Preparación de datos

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
y_train_oh = to_categorical(y_train_enc)
y_test_oh = to_categorical(y_test_enc)

input_dim = X_train_tfidf.shape[1]
tf.random.set_seed(42)

# ---
### Construcción del modelo optimizado

model = Sequential([
    # Primera capa densa con regularización L2 y activación ReLU.
    Dense(256, activation='relu', input_shape=(input_dim,), kernel_regularizer=regularizers.l2(0.001)),
    # Se aumenta el Dropout al 50%.
    Dropout(0.5),
    # Segunda capa densa con regularización L2 y activación ReLU.
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    # Se aumenta el Dropout al 50%.
    Dropout(0.5),
    # Capa de salida.
    Dense(y_train_oh.shape[1], activation='softmax')
])

# ---
### Compilación, entrenamiento y evaluación

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
early = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True) # Aumentamos la paciencia para permitir más épocas de entrenamiento

history = model.fit(X_train_tfidf.toarray(), y_train_oh,
                    validation_split=0.2,
                    epochs=25, # Aumentamos las épocas
                    batch_size=64,
                    callbacks=[early],
                    verbose=1)

# Visualización con Plotly
fig = go.Figure()
fig.add_trace(go.Scatter(y=history.history['loss'], name='Pérdida de Entrenamiento'))
fig.add_trace(go.Scatter(y=history.history['val_loss'], name='Pérdida de Validación'))
fig.update_layout(title='Curvas de Pérdida del Modelo Optimizado', xaxis_title='Época', yaxis_title='Pérdida')
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(y=history.history['accuracy'], name='Exactitud de Entrenamiento'))
fig.add_trace(go.Scatter(y=history.history['val_accuracy'], name='Exactitud de Validación'))
fig.update_layout(title='Curvas de Exactitud del Modelo Optimizado', xaxis_title='Época', yaxis_title='Exactitud')
fig.show()

# Predicciones y reporte
probs = model.predict(X_test_tfidf.toarray())
pred_mlp = le.inverse_transform(np.argmax(probs, axis=1))

print("Accuracy (MLP Optimizado):", round(accuracy_score(y_test, pred_mlp), 3))
print(classification_report(y_test, pred_mlp, digits=3))

# Matriz de Confusión con Plotly
cm = confusion_matrix(y_test, pred_mlp, labels=['negative','neutral','positive'])
labels = ['negative','neutral','positive']
cm_df = go.Figure(data=go.Heatmap(
    z=cm,
    x=labels,
    y=labels,
    colorscale='Blues',
    text=cm,
    texttemplate="%{text}",
    hoverinfo="z"
))
cm_df.update_layout(
    title='Matriz de Confusión – MLP Optimizado (TF‑IDF)',
    xaxis_title='Predicho',
    yaxis_title='Real',
    yaxis_autorange='reversed'
)
cm_df.show()
```
